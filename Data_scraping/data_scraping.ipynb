This function find the number of innings played against each respective team in the tournament.

```python

def get_opposite_teams(href_link):
    page1 = requests.get(href_link)
    soup1 = BeautifulSoup(page1.text, 'html.parser')
    Values1 = soup1.find_all("tr", {"class": "data1"})
    out = []
    for row1 in Values1:
      Values1 = row1.find_all('td', {'class': 'left', 'nowrap': 'nowrap'})

      extracted_texts = []

      for element in Values1:
         a_tags = element.find_all('a')
         for tag in a_tags:
            extracted_texts.append(tag.get_text())
      for text in extracted_texts:
          if "India" in text:
            out.append('India')
          if "Afghanistan" in text:
            out.append('Afghanistan')
          if "Australia" in text:
            out.append('Australia')
          if "Bangladesh" in text:
            out.append('Bangladesh')
          if "England" in text:
            out.append('England')
          if "Netherlands" in text:
            out.append('Netherlands')
          if "New Zealand" in text:
            out.append('New Zealand')
          if "Pakistan" in text:
            out.append('Pakistan')
          if "South Africa" in text:
            out.append('South Africa')
          if "Sri Lanka" in text:
            out.append('Sri Lanka')

    elements_to_count = [
    'Afghanistan', 'Australia', 'Bangladesh', 'England', 'India',
    'Netherlands', 'New Zealand', 'Pakistan', 'South Africa', 'Sri Lanka'
     ]

    element_count = {element: out.count(element) for element in elements_to_count}
    return element_count
```

This function calculate the average runs scored by the player until the previous ODIs played. It is calculated by averaging the 'Avg' column values until the previous ODIs.


```python
def find_avg(col, col_name, Inns):
  x = pd.DataFrame(col)

  x[col_name] = pd.to_numeric(x[col_name], errors='coerce')
  x[col_name].fillna(0, inplace=True)

  # print(type(x.Runs))
  x['Avg_'+col_name] = x[col_name]

  for i in range(1, len(col)):
      x.at[i, 'Avg_'+col_name] = (x.at[i, col_name] + x.at[i - 1, 'Avg_'+col_name]) / 2

  for i in range(1, len(col)):
    # print(type(x.at[i,'Runs']))
    if x.at[i, col_name] == 0:
        x.at[i, 'Avg_'+col_name] = x.at[i-1,'Avg_'+col_name]
  return x['Avg_'+col_name]
```
