```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
req = requests.get('https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338')

soup = bs(req.text, 'html.parser')
```

```python
val = soup.find_all('tbody', {'class' : ''})
```

```python
rows = val[0].find_all('tr')
```

```python
len(rows)
```

```python
matches = {
    'team_1': [],
    'team_2': [],
    'winner': [],
}
```

```python
for row in rows:
    values = row.find_all('span')
    matches['team_1'].append(values[0].text)
    matches['team_2'].append(values[1].text)
    matches['winner'].append(values[2].text)
```

```python
matches_data = pd.DataFrame(matches)
```

```python
matches_data.loc[len(matches_data.index)] = ['India', 'Netherlands', '']
```

```python
matches_data
```

```python
matches_data.loc[matches_data.tail(5).index, 'winner'] = '-'
```

```python
matches_data.to_csv('match_data.csv')
```
